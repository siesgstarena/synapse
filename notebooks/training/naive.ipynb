```python
def EvaluateModel(df,y_pred,x_test):
    for i,prob in enumerate(y_pred):
        curProb = x_test.iloc[i]
        actualProblem = df[(df['C'] == curProb['C'])
        & (df['C++'] == curProb['C++'])
        & (df['Go'] == curProb['Go'])
        & (df['Java'] == curProb['Java'])
        & (df['Javascript'] == curProb['Javascript'])
        & (df['Python'] == curProb['Python'])
        & (df['Python3'] == curProb['Python3'])]['problemId']
        # print(actualProblem.head())
        predProblem = df[df['problemId'] == prob]['problemId']
        diff = np.subtract(df[df.problemId == actualProblem.values[0]].drop(['problemId'], axis=1).values.tolist().pop() , df[df.problemId == predProblem.values[0]].drop(['problemId'], axis=1).values.tolist().pop())
        print(np.sum(diff),np.average(diff))
        
```
