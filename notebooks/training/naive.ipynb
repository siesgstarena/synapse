```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_log_error
from sklearn.metrics import median_absolute_error
from sklearn.metrics import max_error
from sklearn.metrics import explained_variance_score
from sklearn.metrics import mean_poisson_deviance
from sklearn.metrics import mean_gamma_deviance

def EvalM(df,y_pred,y_test):
    y1 = []
    y2 = []
    for i,prob in enumerate(y_pred):
        predProblem = df[df['problemId'] == prob]['problemId']
        actualProblem = df[df['problemId'] == y_test.iloc[i]]['problemId']
        y1.append(np.sum(df[df.problemId == actualProblem.values[0]].drop(['problemId'], axis=1).values.tolist().pop()))
        y2.append(np.sum(df[df.problemId == predProblem.values[0]].drop(['problemId'], axis=1).values.tolist().pop()))
    print(metrics.mean_squared_error(y1, y2))
    print(metrics.r2_score(y1, y2))
    print(metrics.mean_absolute_error(y1, y2))
    print(metrics.median_absolute_error(y1, y2))
    print(metrics.max_error(y1, y2))
    print(metrics.explained_variance_score(y1, y2))
    print(np.std(y1),np.std(y2))
```

```python
EvalM(df,y_pred,y_test)
```

```python
from sklearn.neighbors import NearestNeighbors
def recommend(problemId):
    y_pred = []
    model = NearestNeighbors(n_neighbors=1, algorithm='ball_tree').fit(x_train)
    distances, indices = model.kneighbors(x_train)
    # for i in range(len(distances)):
    #     y_pred.append(y_train.iloc[indices[i][0]])
    for i in range(len(distances)):
        if y_train.iloc[i] == problemId:
            y_pred.append(y_train.iloc[indices[i][0]])
    return y_pred
```
